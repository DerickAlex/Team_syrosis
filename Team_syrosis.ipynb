```python

#importing libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from scipy import stats

%matplotlib inline
```

```python
#Reading the  training Dataset 
train_set = pd.read_csv('train_data.csv')
```

```python
#Brief information regarding the training set 
train_set.info()
```

```python
#checking  unique values
train_set.Weekend.unique()
```

```python
#Assigning the data into a dataframe 
DataFrame1 = pd.DataFrame(train_set)
```

```python
DataFrame1
```

```python
#Dropping the ID column cause it does not affect the prediciton 
DataFrame1 = train_set.drop('ID',1)
DataFrame1
```

```python
#Checking all null values in every column 
DataFrame1.isnull().sum()
```

```python
#Filling out the null values with median value

from sklearn.impute import SimpleImputer

Column_Name = ['Aboutus','Aboutus_Duration','Item_page','Item_page_Duration','Checkout_page','Checkout_page_Duration','Bounce_Rates','Exit_Rates']

for i in Column_Name:
    DataFrame1[Column_Name] =  DataFrame1[Column_Name].fillna( DataFrame1[Column_Name].median())
```

```python
DataFrame1.isnull().sum()
```

```python
#finding object values in each column

for c in DataFrame1.columns:
    
    if DataFrame1[c].dtype == 'object':
        
        Unique_Valu = len(DataFrame1[c].unique())
        print(f'{c} = {Unique_Valu} number of unique values')
```

```python
#Finding the number of occurences of each month  in the dataset

print(DataFrame1['Month'].value_counts().sort_values(ascending=False).head(60))
```

```python
#Replacing these  minimum number of occurences in Month values with 'Others'
MonthName = ['Oct','Sep','Jul','Aug','June','Feb']
for i in MonthName:
    DataFrame1['Month'] =DataFrame1['Month'].replace(i,'Others')
```

```python
print(DataFrame1['Operating_Systems'].value_counts().sort_values(ascending=False).head(60))
```

```python
#Replacing these  minimum number of occurences in OS values with 'Others' 
OS = ['Fedora','CentOS','oracle linux','macOS']
for i in OS:
    DataFrame1['Operating_Systems'] = DataFrame1['Operating_Systems'].replace(i,'Others')
```

```python
#Converting all categorical values into nominal values
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
EncodeColumns = ['Month','Operating_Systems','VisitorType','Weekend']
for i in EncodeColumns:
    DataFrame1[i] = LE.fit(DataFrame1[i]).transform(DataFrame1[i])
```

```python
#Checking the correlation
DataFrame1.corr()
```

```python
DataFrame1.info()
```

```python
DataFrame1
```

```python
#Converting Categorical columns into numeric columns via OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
Hot = OneHotEncoder()
HotEncoder = Hot.fit(DataFrame1[['Month','Operating_Systems','VisitorType','Weekend']]).transform(DataFrame1[['Month','Operating_Systems','VisitorType','Weekend']])
```

```python
Temp = pd.DataFrame(HotEncoder.toarray(), columns=Hot.get_feature_names() )
```

```python
Temp
```

```python
DataFrame2 = pd.concat([Temp, DataFrame1[['Aboutus','Aboutus_Duration','Item_page','Item_page_Duration','Checkout_page','Checkout_page_Duration','Bounce_Rates','Exit_Rates','Page_Values','Remarkable_Day','Browser','Province','TrafficType','Income']]], axis=1, join='inner')
```

```python
DataFrame2
```

```python
#Choosing only the dependent variables for X
X = DataFrame2.drop('Income',1)
#Choosing only the independent variable for Y
Y = DataFrame2['Income']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30)
```

```python

from sklearn.ensemble import RandomForestClassifier
Classifier = RandomForestClassifier(n_estimators = 250, max_depth=16,  criterion ='gini' )
Classifier.fit(X_train, Y_train)
```

```python
Y_pred = Classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
Result = accuracy_score(Y_test,Y_pred)
print("Accuracy:",Result)
```

```python
from sklearn.svm import SVC
Model = SVC(kernel= 'linear')
```

```python
Model.fit(X_train,Y_train)
```

```python
Model_Score = Model.score(X_test,Y_test)
```
